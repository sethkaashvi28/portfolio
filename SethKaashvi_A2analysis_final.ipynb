**<h1>Introduction**

In San Francisco and the Bay Area, where Karl the Fog rolls in and out and the rising cost of transportation presents daily challenges residents and visitors alike are seeking more sustainable and flexible commuting options. Muni, BART and bike sharing programs offer eco friendly alternatives to traditional transportation methods. Bay Wheels a leading provider of shared bicycles in the Bay Area plays a critical role in offering an affordable, convenient and environmentally conscious solution to mobility. Bay Wheels network of bikes makes it easy for people to navigate the city and its surrounding areas, providing an essential service for reducing carbon emissions and enhancing last-mile connectivity.

As interest in sustainable urban mobility continues to grow, understanding user behavior is crucial for making data-driven business decisions. This analysis uses Bay Wheels rental data to uncover key trends related to usage patterns across hours of the day, weather conditions, and seasonal changes with the goal of providing actionable insights that can help optimize operations and improve service delivery for users.


```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# dataset uploaded
file = ('/content/sample_data/bikes_data.xlsx')

df = pd.read_excel(io = file,
                  sheet_name = 0,
                  header = 0)
#checking dataset
print(df.head(n=10))
```

```python
#missing value
print(df.isnull().sum())
```

```python
print(df.describe())
```

```python
#Checking hours of the day with average number of rentals

# Set the size of the plot (width = 12 inches, height = 4 inches)
plt.figure(figsize=(12, 4))

# Create a bar plot with 'Hour' on the x-axis and 'RENTALS' on the y-axis to show rental patterns across hours of the day
sns.barplot(data=df, x='Hour', y='RENTALS',  color='orange')

# Set the title of the plot to reflect that peak rentals occur during commuting hours
plt.title('Commuting Hours Drive Peak Bike Rentals')

# Label the x-axis as 'Hour of Day'
plt.xlabel('Hour of Day')

# Label the y-axis as 'Average Number of Rentals'
plt.ylabel('Average Number of Rentals')

# Display the plot
plt.show()
```

This code creates a bar plot that visualizes how bike rentals vary across different hours of the day. It highlights peak rental times during the morning and evening commute hours. The plot helps identify key usage patterns, suggesting that the majority of bike rentals occur during weekday commuting hours, which is crucial for optimizing Bay Wheels' operations around peak demand times.

**Why Are the Majority of Users Renting Bicycles?**

The hourly rental data for Bay Wheels shows clear peaks around 8am and 5-6pm which align with common work commute times. These patterns indicate that the majority of users are renting bikes primarily for weekday commuting rather than for casual or recreational use. The demand during these peak hours suggests that many individuals depend on Bay Wheels as a reliable mode of transportation for their daily work commute, especially given the rising cost of traditional transportation options like Muni and BART in the Bay Area. The lower number of rentals during midday and nighttime further reinforces this commuter-driven usage pattern.


```python
#Checking if it rains what is the average number of rentals

#Set the figure size for better clarity
plt.figure(figsize=(6, 4))

# Create a bar plot with the x-axis representing whether it rained or not (True/False)
# and the y-axis representing the average number of rentals
sns.barplot(data=df, x=df['Rainfall'] > 0, y='RENTALS', palette = 'pastel')

# Set the title for the plot
plt.title('How Rain Affects Bike Usage')

# Label the x-axis to indicate whether it rained (True = Yes, False = No)
plt.xlabel('Did It Rain? (True = Yes, False = No)')

# Label the y-axis to show the average number of rentals
plt.ylabel('Average Number of Rentals')

# Display the plot
plt.show()
```

This code generates a bar plot to visualize how rainfall affects bike rental demand. The x-axis shows whether it rained (True/False), and the y-axis represents the average number of rentals for each condition. The plot helps to reveal if rainy weather significantly reduces bike usage compared to dry days.

**Are There Ideal Weather Conditions That Create High Bike Rental Demand?**

Looking at the data we can see a clear pattern when it rains bike rentals take a significant dip. The bar plot comparing rentals on rainy days to those on dry days highlights just how much weather impacts usage. On rainy days the number of rentals drops sharply, which makes sense who wants to ride a bike when it's wet and uncomfortable. On dry days, the average number of rentals is much higher, suggesting that people are more likely to hop on a bike when the weather is clear and pleasant.

This finding shows that clear, dry weather is ideal for bike rentals. Rainy conditions are a major deterrent probably because of concerns over safety, visibility, and the overall discomfort of riding in the rain. Given this, Bay Wheels should plan ahead for rainy days by adjusting their operations.

```python
#Checking seasons with average number of rentals

# Set the size of the plot
plt.figure(figsize=(6, 4))

# Create a bar plot showing the average number of rentals by season
sns.barplot(data=df, x='Season', y='RENTALS', color = 'yellow')

# Set the title of the plot to indicate it shows seasonal rental trends
plt.title('Seasonal Trends in Bike Rentals')

# Label the x-axis to indicate the seasons
plt.xlabel('Season')

# Label the y-axis to indicate the average number of rentals
plt.ylabel('Average Number of Rentals')

# Display the plot
plt.show()
```

This code generates a bar plot to visualize bike rental trends across different seasons. The x-axis represents the seasons, and the y-axis shows the average number of rentals in each season. By analyzing this plot we can observe seasonal patterns in bike usage helping identify which times of the year see the highest demand for rentals.

**How Should the Company Strategize Based on These Results?**

The seasonal trends in bike rentals are clear Summer sees the highest demand followed by Autumn, Spring, and then Winter. The "Seasonal Trends in Bike Rentals" bar plot below illustrates this perfectly showing that bike rentals spike during warmer months and dip in the colder months. This pattern is likely driven by factors such as favorable weather, longer daylight hours and increased outdoor activity during the warmer seasons. People are naturally more inclined to rent bikes when the weather is pleasant which is reflected in the rental data.


With these seasonal shifts in mind, Bay Wheels should implement a strategic approach to demand planning. For example, as we move into Summer and Autumn, the company should increase bike availability and maintenance efforts to ensure there are enough bikes in circulation during peak rental periods. In contrast, Winter months may require lighter resource allocation due to the dip in demand, but this could also be an opportunity for Bay Wheels to introduce seasonal promotions or discounts to maintain rider engagement. By aligning resources with seasonal trends, the company can optimize operations, improve efficiency, and increase customer satisfaction by ensuring bikes are available when users need them most.
